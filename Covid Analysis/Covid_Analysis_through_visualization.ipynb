```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.DataFrame()
```

```python
if data.shape[0] == 0:
    url = "https://github.com/owid/covid-19-data/blob/master/public/data/owid-covid-data.csv?raw=true"
    data = pd.read_csv(url)
```

```python
data.head()
```

```python
data[['iso_code','location']].drop_duplicates().tail(25)
```

```python
data['date'] = pd.to_datetime(data['date'], format='%Y-%m-%d')
```

```python
# Set the cc_code to either 1: India, 2: Netherlands, 3: USA or 4: Great Britain
cc_code = 3
```

```python
if cc_code == 1:
  iso_code = 'IND'
  location = 'India'
elif cc_code == 2:
  iso_code = 'NLD'
  location = 'Netherlands'
elif cc_code == 3:
  iso_code = 'USA'
  location = 'United States'
elif cc_code == 4:
  iso_code = 'GBR'
  location = 'United Kingdom'
```

```python
df = data[data['location'] == location]
```

```python
plt.figure(figsize = (15,9))
sns.lineplot(data = df,x='date', y='new_cases')
sns.lineplot(data = df,x='date', y='new_cases_smoothed')
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.title('New Cases in '+ location)
plt.show()
```

```python
plt.figure(figsize = (15,9))
sns.lineplot(data = df,x='date', y='new_deaths')
sns.lineplot(data = df,x='date', y='new_deaths_smoothed')
plt.xlabel('Date')
plt.ylabel('New Deaths')
plt.title('New Deaths in '+ location)
plt.show()
```

```python
plt.figure(figsize = (15,9))
sns.lineplot(data = df,x='date', y='total_cases')
#sns.lineplot(data = df,x='date', y='total_cases_smoothed')
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.title('Total Cases in '+ location)
plt.show()
```

```python
plt.figure(figsize = (15,6))
sns.lineplot(data = df,x='date', y='new_vaccinations')
sns.lineplot(data = df,x='date', y='new_vaccinations_smoothed')
plt.xlabel('Date')
plt.ylabel('New Vaccinations')
plt.title('New Vaccinations in '+ location)
plt.show()
```

```python
plt.figure(figsize = (15,9))
sns.lineplot(data = df,x='date', y='total_vaccinations')
#sns.lineplot(data = df,x='date', y='total_cases_smoothed')
plt.xlabel('Date')
plt.ylabel('Total Vaccinations')
plt.title('Total Vaccinations in '+ location)
plt.show()
```
